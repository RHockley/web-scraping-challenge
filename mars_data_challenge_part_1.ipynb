```python
#Import Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
from pprint import pprint
import pymongo
```

```python
# Set up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Initialise PyMongo
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.mars_news_db
collection = db.articles
```

```python
# Mars News Website to be scraped
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
# Iterate
for x in range(20):
    # Define HTML object
    html = browser.html
    # Parse HTML using Beautiful Soup
    soup = BeautifulSoup(html, 'html.parser')
    # Retrieve all elements with information
    elements = soup.find_all('div', class_='list_text')
    
        # Iterate through each one
    for element in elements:
        
        # Scrape the title 
        title = element.find('div', class_='content_title').text.strip()
        
        # Scrape the preview
        preview = element.find('div', class_='article_teaser_body').text.strip()
        
        print('___________________________________________________________________________________________________')
        print(title)
        print(preview)
        
        
        # Dictionary for MongoDB
        data = {
            'title': title,
            'preview': preview,
        }
        collection.insert_one(data)
```

```python
# Display records as a Python list
ls_dict=[]
items = db.articles.find()
for item in items:
    ls_dict.append(item)
    
pprint(ls_dict)
```

```python
browser.quit()
```
